```python
# 25. –°—Ç–≤–æ—Ä–∏ –¥–≤—ñ –º–Ω–æ–∂–∏–Ω–∏:

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# —ñ –∑–Ω–∞–π–¥–∏:  
# - –æ–±‚Äô—î–¥–Ω–∞–Ω–Ω—è (union)
# - –ø–µ—Ä–µ—Ç–∏–Ω (intersection)
# - —Ä—ñ–∑–Ω–∏—Ü—é (difference)

print(A | B)
print(A ^ B)
print(A - B)
```
